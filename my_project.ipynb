```python
#Importing the Necessary Libraries for Analysis

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Conclusions
#The data set had no missing values.
#Changed the life expectancy at birth -> LE. This was done for ease of use
#There is a positive corelation between GDP and LE for every country.
#There was a slight recession (GDP) during the years 2008 and 2009.
#Zimbabwe had the largest change in both GDP and LE, which were both positivly correlated.
```

```python
#Exploring initial data set

data = pd.read_csv('all_data.csv')
data.head(20)
```

```python
# 1. Cleaning and Preprocessing

# Checking for missing values 

print(data.isnull().sum())
```

```python
# Renaming 'Life expectancy at birth' to LE
# For ease of use.

data.rename(columns={'Life expectancy at birth (years)':'LE'}, inplace=True) #inplace does not create new DF

data.head()
```

```python
# Exploring GDP for each country using boxplots. 
# Observing the range of values, and percentiles, for each country

plt.figure(figsize=(11, 6))
sns.boxplot(data = data, x = 'Country', y = 'GDP')

```

```python
# Exploring each countries GDP change over time using line plots

plt.figure(figsize = (11, 6))
sns.lineplot(data=data, x='Year', y='GDP', hue="Country")
plt.show()
plt.clf()

# Analyzing each countries GDP Change on it's own scale 
```

```python
#Analyzing GDP versus LE for each country

countries = ['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe']
#create a list of each country

#creating subplots

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
#first creates a canvas using figsize, then splits into 2 rows by 3 columns. Returns figure and individual subplots

axes = axes.flatten()

for i, country in enumerate(countries):
    
    country_data = data[data['Country'] == country]
    
    sns.scatterplot(x='GDP', y='LE', data=country_data, ax=axes[i])  
    
    sns.regplot(x='GDP', y='LE', data=country_data, scatter=False, ax=axes[i])
    
    axes[i].set_title(f'{country} GDP vs. Life Expectancy')
    
    
plt.tight_layout()
    
plt.show()

#These plots show that there is a positive correlation between an increase in GDP and Life Expectancy.



```

```python
#Exploring the Life Expectancy for each country

plt.figure(figsize=(11, 6))
sns.boxplot(data = data, x = 'Country', y = 'LE')

#This figure tells us many things about the distribution of age.

#The average age in zimbabwe is less than all countries by a substantial margin
#The avarage ages for Chile and Germany surpass the United States, which has the highest GDP by far. 
```

```python
#Analyzing GDP change over time for each country individually

countries = ['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe']
#create a list of each country

#creating subplots

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
#first creates a canvas using figsize, then splits into 2 rows by 3 columns. Returns figure and individual subplots

axes = axes.flatten()

for i, country in enumerate(countries):
    
    country_data = data[data['Country'] == country]
    
    sns.lineplot(x='Year', y='GDP', data=country_data, ax=axes[i])  
    
    axes[i].set_title(f'{country} GDP Over time for each country')
    
    
plt.tight_layout()
    
plt.show()

#The purpose of these graphs is to analyze each individuals GDP growth over time. 
#What these graphs tell us is that there there was a recession around 2008 - 2009.  
```

```python
#Analyzing LE over time for each country

countries = ['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe']
#create a list of each country

#creating subplots

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
#first creates a canvas using figsize, then splits into 2 rows by 3 columns. Returns figure and individual subplots

axes = axes.flatten()

for i, country in enumerate(countries):
    
    country_data = data[data['Country'] == country]
    
    sns.lineplot(x='Year', y='LE', data=country_data, ax=axes[i])  
    
    axes[i].set_title(f'{country} Life Expectancy over time for each country')
    
    
plt.tight_layout()
    
plt.show()

#Less developed countries such as Chile and Mexico show a slight drop in LE during the Great Recession. 
#However, developed countries show little to no change in LE over time. Zimbabwe has had a sharp increase in LE over time.
```
